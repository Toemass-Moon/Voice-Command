```python
import speech_recognition as sr
from word2number import w2n
import pyautogui as pag
import time
import os
```

```python
on = True

while on:
    recog = sr.Recognizer()
    
    with sr.Microphone() as source:
        audio = recog.listen(source)

        try:
            text = recog.recognize_google(audio)
        except:
            text = None 
    if text == 'turn off':
        print('Turning off Voice Command')
        on = False
        
    elif text == 'computer': 
        og_x, og_y = pag.position()
        
        with sr.Microphone() as source:
            print('Talk Now')
            audio = recog.listen(source)

            try:
                text = recog.recognize_google(audio)
                print(f'You said: {text}')
            except:
                print('Nothing was recognized')

        # will open said website on a new tab
        if text.split()[0] == 'open':

            new_tab_1 = pag.locateOnScreen('./notebook_imgs/new_tab.png')
            new_tab_2 = pag.locateOnScreen('./notebook_imgs/new_tab_2.png')
            new_tab_3 = pag.locateOnScreen('./notebook_imgs/new_tab_other_window.png')

            try:
                x, y, _, _ = new_tab_1
            except:
                try:
                    x, y, _, _ = new_tab_2
                except:
                    x, y, _, _ = new_tab_3

            pag.moveTo(x + 50, y + 16)
            pag.click()
            pag.write(text.split()[-1])
            pag.press('enter')
            pag.moveTo(og_x, og_y)

        # will open said website on the current tab
        elif 'go to' in text:
            search = pag.locateOnScreen('./notebook_imgs/search_bar.png')
            x, y, _, _ = search

            pag.moveTo(x + 50, y + 15)
            pag.click()
            pag.write(text.split()[-1])
            pag.press('enter')
            pag.moveTo(og_x, og_y)
            
        # will minimize the window
        elif text == 'minimize window':
            minimize_1 = pag.locateOnScreen('./notebook_imgs/minimize.png')
            minimize_2 = pag.locateOnScreen('./notebook_imgs/minimize_other_window.png')

            try:
                x, y, _, _ = minimize_1
            except:
                x, y, _, _ = minimize_2

            pag.moveTo(x + 12, y + 7)
            pag.click()
            pag.write(text.split()[-1])
            pag.press('enter')
            pag.moveTo(og_x, og_y)
            
        # Take a screenshot
        elif text == 'screenshot':
            pag.screenshot(f'./Screenshots/{str(int(time.time()))}.png')
        
        # Opening League
        elif text == 'League of Legends':
            os.startfile('C:\\Riot Games\\League of Legends\\LeagueClient.exe')
            
        # closeing the browser Window
        elif text == 'close window':
            close_1 = pag.locateOnScreen('./notebook_imgs/close_window.png')
            close_2 = pag.locateOnScreen('./notebook_imgs/close_other_window.png')
            close_3 = pag.locateOnScreen('./notebook_imgs/close_other_window_2.png')
            try:
                x, y, _, _ = close_1 
            except: 
                try:
                    x, y, _, _ = close_2
                except:
                    x, y, _, _ = close_3

            pag.moveTo(x + 8, y + 7)
            pag.click()
            pag.write(text.split()[-1])
            pag.press('enter')
            pag.moveTo(og_x, og_y)
        
        # closing the current tab
        elif text == 'close tab':
            current_tab = pag.locateOnScreen('./notebook_imgs/current_tab.png')

            try:
                x, y, _, _ = current_tab
            except:
                print('could not find image')

            pag.moveTo(x+7, y+15)
            pag.click()
            pag.moveTo(og_x, og_y)
            
        # close last tab
        elif text == 'close last tab':
            close_tab_1 = pag.locateOnScreen('./notebook_imgs/new_tab.png')
            close_tab_2 = pag.locateOnScreen('./notebook_imgs/new_tab_2.png')
            close_tab_3 = pag.locateOnScreen('./notebook_imgs/new_tab_other_window.png')

            try:
                x, y, _, _ = close_tab_1
            except:
                try:
                    x, y, _, _ = close_tab_2
                except:
                    x, y, _, _ = close_tab_3


            pag.moveTo(x + 13, y+16)
            pag.click()
            pag.moveTo(og_x, og_y)
        
        # volume up
        elif "volume up" in text:
            try:
                num = w2n.word_to_num(text.split()[-1])
                
            except:
                num = int(text.split()[-1])
                
            pag.press('volumeup', presses = num)
            
        # volume down
        elif "volume down" in text:
            try:
                num = w2n.word_to_num(text.split()[-1])
                
            except:
                num = int(text.split()[-1])
                
            pag.press('volumedown', presses = num)
```
